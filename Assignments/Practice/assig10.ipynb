```python
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer,WordNetLemmatizer
from nltk.corpus import stopwords
from nltk import pos_tag
import nltk
```

```python
with open("../../Datasets/file1.txt") as file:
    doc1=file.read()
doc1
```

```python
with open("../../Datasets/file2.txt") as file:
    doc2=file.read()
doc2
```

```python
with open("../../Datasets/file3.txt") as file:
    doc3=file.read()
doc3
```

```python
docs=[doc1,doc2,doc3]
```

```python
tokens={}
i=0

for doc in docs:
    tokens[i]=word_tokenize(doc)
    i+=1
tokens[0]
```

```python
stop_words=set(stopwords.words('english'))
```

```python
stop_words
```

```python
filtered_tokens={}
i=0

for token in tokens:
    filtered_tokens[i]=[t for t in tokens[token] if t.lower() not in stop_words]
    i+=1
filtered_tokens[0]
```

```python
stemmer=PorterStemmer()
stemmed_tokens={}
i=0

for filtered_token in filtered_tokens:
    stemmed_tokens[i]=[stemmer.stem(token) for token in filtered_tokens[filtered_token]]
    i+=1
stemmed_tokens[0]
```

```python
lemma=WordNetLemmatizer()
lemma_tokens={}
i=0

for filtered_token in filtered_tokens:
    lemma_tokens[i]=[lemma.lemmatize(token) for token in filtered_tokens[filtered_token]]
    i+=1
    
lemma_tokens[0]
```

```python
postags={}
i=0

for filtered_token in filtered_tokens:
    postags[i]=pos_tag(filtered_tokens[filtered_token])
postags[0]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer=TfidfVectorizer()
tfidf_matrix=vectorizer.fit_transform(docs)

feature_names=vectorizer.get_feature_names_out()

for i in range(len(docs)):
    print("Document : ",docs[i])
    for j,word in enumerate(feature_names):
        tfidf_value=tfidf_matrix[i,j]
        print("word : ",word," tfidf : ",tfidf_value)
```

```python
tf={}

for filtered_token in filtered_tokens[0]:
    if(filtered_token in tf):
        tf[filtered_token]+=1
    else:
        tf[filtered_token]=1
       
word_counts=len(filtered_tokens[0])

for tf_word in tf:
    tf[tf_word]=tf[tf_word]/word_counts
    print(tf_word," : ",tf[tf_word])

```

```python
import numpy as np
tfidf={}

for word in filtered_tokens[1]:
    for filtered_token in filtered_tokens:
        if(word in filtered_tokens[filtered_token]):
            if(word in tfidf):
                tfidf[word]+=1
            else:
                tfidf[word]=1
    tfidf[word]=np.log(3/tfidf[word])
    print(word," : ",tfidf[word])
        
```
