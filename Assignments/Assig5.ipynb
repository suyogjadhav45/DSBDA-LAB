```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("../Datasets/StudentsPerformance.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df["math score"].fillna(df["math score"].mean(),inplace=True)
df["reading score"].fillna(df["reading score"].mean(),inplace=True)
df["writing score"].fillna(df["writing score"].mean(),inplace=True)
df
```

```python
df.isnull().sum()
```

```python
# Group by categorical variable and calculate statistics like mean median etc for numeric variables
# categrical variable - gender
# numeric variable - math score

stats_gender_math=df.groupby('gender')['math score'].agg(['mean','median','min','max','std'])
print(stats_gender_math)
```

```python
# categrical variable - race/ethnicity
# numeric variable - math score

stats_race_math=df.groupby('race/ethnicity')['math score'].agg(['mean','median','min','max','std'])
print(stats_race_math)

stats_race_math_list=stats_race_math.values.flatten().tolist()
stats_race_math_list
```

```python
df2=pd.read_csv("../Datasets/iris.csv")
```

```python
# Finding mean by grouping intervals
id_interval=[0,50,100,150]
id_labels=["0-50","51-100","101-150"]

df2["idset"]=pd.cut(df2["Id"],bins=id_interval,labels=id_labels,right=False)
summery_stats_id=df2.groupby("idset")["SepalLengthCm"].agg(["mean","median","min","max"])
summery_stats_id
```
